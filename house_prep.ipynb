```python
from sklearn.preprocessing import StandardScaler

X = df.drop("MEDV", axis=1)

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
processed_df = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

processed_df["MEDV"] = df["MEDV"]

processed_df.head()
```

```python
processed_df.to_csv(
    "processed_house_data.csv",
    index=False
)
```
